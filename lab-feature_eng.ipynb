There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship=spaceship.dropna()
```

```python
spaceship.isna().sum()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship['Cabin'].unique()
```

```python
spaceship['Cabin'] = spaceship['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else np.nan)
```

```python
spaceship['Cabin'].unique()
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(['PassengerId','Name'], axis=1)
```

```python
spaceship.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['HomePlanet', 'Cabin', 'Destination', 'CryoSleep', 'VIP']

cat = [spaceship[col].unique() for col in categorical_cols]

ohe = OneHotEncoder(sparse_output=False, categories=cat)

ohe.fit(spaceship[categorical_cols]) 
spaceship_trans_np = ohe.transform(spaceship[categorical_cols])

spaceship_trans_np
```

```python
spaceship2 = pd.DataFrame(spaceship_trans_np, columns=ohe.get_feature_names_out(), index=spaceship.index)
spaceship2
```

```python
spaceship2.dtypes
```

```python
#split numerical values of spaceship
#concat numerical df and spaceship2
#define x and y
#train test split
```

```python
numerical_df = spaceship.select_dtypes(include=['number'])
```

```python
concatenated_df = pd.concat([spaceship2, numerical_df], axis=1)
```

**Perform Train Test Split**

```python
X = concatenated_df
y = spaceship['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
knn.fit(X_train, y_train)
```

- Evaluate your model's performance. Comment it

```python
pred = knn.predict(X_test)
pred
```

```python
y_test.values
```

```python
print(f"The accuracy of the model is {knn.score(X_test, y_test)*100: .2f}%")
```
